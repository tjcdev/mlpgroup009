```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Baseline

#### Bipedal Walker (base-1)

```python
base_bip = pd.read_json('base-bip-10m.json')
sns_plot = sns.lineplot(x="nupdates", y="eprewmean",
             ci="explained_variance",
             err_style="band",
             data=base_bip)
plt.ylim(-160, 300)
plt.show()
sns_plot.get_figure().savefig("baseline_normal.png",  bbox_inches = 'tight')
```

#### Biphard (base-2)

```python
base_biphard = pd.read_json('base-biphard-10m.json')
sns_plot = sns.lineplot(x="nupdates", y="eprewmean", data=base_biphard)
plt.ylim(-160, 50)
plt.show()

sns_plot.get_figure().savefig("baseline_hardcore.png",  bbox_inches = 'tight')
```
